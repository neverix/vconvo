```python
import librosa
import librosa.display
import numpy as np
```

```python
y, sr = librosa.load("aaaaa.wav")
```

```python
s = librosa.stft(y)
```

```python
from IPython.display import Audio

def play(s):
    return Audio(librosa.istft(s), rate=sr)

play(s)
```

```python
def show(s):
    librosa.display.specshow(librosa.amplitude_to_db(s, ref=np.max), y_axis="log")

show(s)
```

```python
from matplotlib import pyplot as plt

plt.plot(np.abs(librosa.amplitude_to_db(s, ref=np.max).T[100]))
```

```python
pitches, magnitudes = librosa.core.piptrack(S=s, sr=sr, fmin=75, fmax=160)
```

```python
show(pitches)
```

```python
pitch = pitches.argmax(axis=0)
plt.plot(pitch)
```
