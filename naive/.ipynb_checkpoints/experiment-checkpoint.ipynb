```python
import librosa
import librosa.display
import numpy as np
```

```python
from IPython.display import Audio
```

```python
y, sr = librosa.core.load("./me.wav")
```

```python
Audio(y, rate=sr)
```

```python
n_fft=2048
hop_length=512
```

```python
d = librosa.stft(y, n_fft=n_fft)
d[:8, :] = 0
```

```python
def show(d):
    librosa.display.specshow(d, sr=sr, x_axis='time', y_axis='log')
show(d)
```

```python
def show_t(d):
    librosa.display.specshow(d, sr=sr, x_axis='linear', y_axis='time')
show_t(d.T)
```

```python
def play(d):
    return Audio(librosa.istft(d), rate=sr)
play(d)
```
